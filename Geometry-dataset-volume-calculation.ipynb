```python
import numpy as np

gdf = volumes_filtered.copy()
gdf["geometry"] = gdf.geometry.buffer(0)
gdf = gdf.reset_index(drop=True)

sidx = gdf.sindex
geoms = gdf.geometry.values
funcs = gdf["function"].to_numpy()   # <-- column to enforce same-value merging
n = len(gdf)

parent = np.arange(n, dtype=int)
rank = np.zeros(n, dtype=int)

def find(x):
    while parent[x] != x:
        parent[x] = parent[parent[x]]
        x = parent[x]
    return x

def union(a, b):
    ra, rb = find(a), find(b)
    if ra == rb:
        return
    if rank[ra] < rank[rb]:
        parent[ra] = rb
    elif rank[ra] > rank[rb]:
        parent[rb] = ra
    else:
        parent[rb] = ra
        rank[ra] += 1

for i, gi in enumerate(geoms):
    for j in sidx.intersection(gi.bounds):
        if j <= i:
            continue

        # Reject merge if function values differ (also handles NaNs safely)
        if funcs[i] != funcs[j]:
            continue

        if gi.touches(geoms[j]):
            union(i, j)

gdf["_cluster"] = [find(i) for i in range(n)]

# dissolve within each cluster (function will be constant within cluster now)
merged = gdf.dissolve(by="_cluster").reset_index(drop=True)

merged["area_m2"] = merged.geometry.area
merged["volume_m3"] = merged["area_m2"] * merged["measHeight"]

merged.head()
```

```python
import numpy as np
import geopandas as gpd

def merge_overlaps_keep_attrs(
    gdf: gpd.GeoDataFrame,
    area_col="area_m2",
    name_col="Name",
    function_col="function",      # <-- NEW: attribute to restrict merging
    merge_nan_functions=False     # <-- NEW: if True, NaN merges with NaN
):
    gdf = gdf.copy()
    gdf["geometry"] = gdf.geometry.buffer(0)   # fix invalids
    gdf = gdf.reset_index(drop=True)

    sidx = gdf.sindex
    geoms = gdf.geometry.values
    n = len(gdf)

    # cache function values for fast comparison
    funcs = None
    if function_col is not None and function_col in gdf.columns:
        funcs = gdf[function_col].to_numpy()

    parent = np.arange(n, dtype=int)
    rank = np.zeros(n, dtype=int)

    def find(x):
        while parent[x] != x:
            parent[x] = parent[parent[x]]
            x = parent[x]
        return x

    def union(a, b):
        ra, rb = find(a), find(b)
        if ra == rb:
            return
        if rank[ra] < rank[rb]:
            parent[ra] = rb
        elif rank[ra] > rank[rb]:
            parent[rb] = ra
        else:
            parent[rb] = ra
            rank[ra] += 1

    def same_function(i, j):
        if funcs is None:
            return True
        fi, fj = funcs[i], funcs[j]

        # normal equality (covers strings/ints/etc.)
        if fi == fj:
            return True

        # optional: treat NaN == NaN for merging
        if merge_nan_functions:
            try:
                return np.isnan(fi) and np.isnan(fj)
            except TypeError:
                return False

        return False

    # cluster by TRUE overlap only (not touches), AND same function value
    for i, gi in enumerate(geoms):
        for j in sidx.intersection(gi.bounds):
            if j <= i:
                continue

            # Reject merge early if function differs
            if not same_function(i, j):
                continue

            gj = geoms[j]
            if gi.intersects(gj):
                inter = gi.intersection(gj)
                if (not inter.is_empty) and (inter.area > 0):
                    union(i, j)

    gdf["_cluster"] = [find(i) for i in range(n)]

    out_rows = []
    for _, grp in gdf.groupby("_cluster", sort=False):
        merged_geom = grp.geometry.unary_union

        # take attributes from biggest polygon
        rep = grp.loc[grp[area_col].idxmax()].copy()

        # Name rule
        if name_col in grp.columns:
            names = [x for x in grp[name_col].tolist()
                     if x is not None and str(x).strip() != ""]
            uniq = list(dict.fromkeys(names))  # unique, preserve order
            rep[name_col] = None if len(uniq) == 0 else (uniq[0] if len(uniq) == 1 else uniq)

        rep["geometry"] = merged_geom
        out_rows.append(rep)

    out = gpd.GeoDataFrame(out_rows, crs=gdf.crs).reset_index(drop=True)
    return out

# usage
merged_no_overlaps = merge_overlaps_keep_attrs(
    merged,
    area_col="area_m2",
    name_col="Name",
    function_col="function"   # change if your column name differs
)
```
