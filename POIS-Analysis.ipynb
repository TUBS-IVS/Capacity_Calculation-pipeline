```python
import geopandas as gpd

pois_gpkg = r"Areas-of-interest-POIs\ALL_POIS.gpkg"

# Load the POI layer
pois = gpd.read_file(pois_gpkg, layer="pois")

print(f"Loaded {len(pois)} POIs")
pois.head()
```

```python
with open("Areas-of-interest-POIs/ALL_POIS_COLUMNS.txt", "w", encoding="utf-8") as f:
    for col in pois.columns:
        f.write(col + "\n")

print("Saved ALL columns to ALL_POIS_COLUMNS.txt")
```
