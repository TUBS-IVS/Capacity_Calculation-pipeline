```python
import numpy as np

addr_cols = [
    "addr:housename",
    "addr:street",
    "addr:housenumber",
    "addr:place",
    "addr:city",
    "addr:postcode",
]

def merge_address(row):
    parts = [
        str(row[c]).strip()
        for c in addr_cols
        if c in row
        and pd.notna(row[c])
        and str(row[c]).strip() != ""
        and str(row[c]).lower() != "nan"
    ]
    return np.nan if not parts else ", ".join(parts)

pois["addr:full"] = pois.apply(merge_address, axis=1)
```

```python
pois = pois.drop(columns=addr_cols)
pois.head()
```

```python
pois["name"] = pois["name"].fillna(pois["operator"])
pois = pois.drop(columns=["operator"], errors="ignore")
pois.head()
```

```python
pois["website"] = (
    pois["website"]
    .replace("", np.nan)
    .fillna(pois["url"])
)

pois = pois.drop(columns=["url"], errors="ignore")
pois.head()
```

```python
import ast

def ensure_dict(x):
    if isinstance(x, dict):
        return x
    if isinstance(x, str) and x.strip():
        try:
            v = ast.literal_eval(x)
            return v if isinstance(v, dict) else {}
        except Exception:
            return {}
    return {}

pois["tags_dict"] = pois["tags"].apply(ensure_dict)

def dict_to_tags_text(d: dict) -> str:
    if not d:
        return np.nan

    items = []
    for k, v in d.items():
        if v is None or (isinstance(v, float) and np.isnan(v)):
            continue
        ks = str(k).strip()
        vs = str(v).strip()
        if ks == "" or vs == "" or vs.lower() == "nan":
            continue

        # keep both "key=value" and also the raw value for easier searching
        items.append(f"{ks}={vs}")

    return np.nan if not items else "; ".join(items)

pois["tags_text"] = pois["tags_dict"].apply(dict_to_tags_text)
```

```python
import re

def ensure_dict(x):
    if isinstance(x, dict):
        return x
    if isinstance(x, str) and x.strip():
        try:
            v = ast.literal_eval(x)
            return v if isinstance(v, dict) else {}
        except Exception:
            return {}
    return {}

pois["tags_dict"] = pois["tags"].apply(ensure_dict)

_non_alnum = re.compile(r"[^0-9a-zA-ZäöüÄÖÜß]+")  # keep DE chars too
_ws = re.compile(r"\s+")

def tokenize_text(s: str) -> str:
    s = str(s).lower()
    s = _non_alnum.sub(" ", s)   # replace EVERYTHING non-alnum with space
    s = _ws.sub(" ", s).strip()  # collapse spaces
    return s

def tags_to_searchable_text(d: dict) -> str:
    if not isinstance(d, dict) or not d:
        return np.nan

    # combine keys + values into one big string, then tokenize once
    raw = " ".join([f"{k} {v}" for k, v in d.items() if pd.notna(v)])
    out = tokenize_text(raw)

    return np.nan if out == "" else out

pois["tags_search"] = pois["tags_dict"].apply(tags_to_searchable_text)
```

```python
base_cols = [
    'id', 'email', 'name', 'website', 'amenity', 'building',
    'shop', 'tourism', 'information', 'geometry',
    'addr:full', 'tags_search'
]

merge_cols = [
    'landuse', 'bar', 'bbq', 'biergarten',
    'building:levels', 'cafe', 'college', 'fast_food', 'fuel',
    'gambling', 'ice_cream', 'office', 'post_office',
    'restaurant', 'social_facility', 'wikipedia', 'books', 'clothes',
    'coffee', 'craft', 'furniture', 'hairdresser', 'kiosk', 'lottery',
    'massage', 'medical_supply', 'organic', 'religion', 'second_hand',
    'shoes', 'tobacco', 'trade', 'vacuum_cleaner', 'wholesale',
    'guest_house', 'museum',
    'kindergarten', 'police', 'school', 'kitchen', 'laundry', 'hotel'
#     'osm_type',
]

def consolidate_info(row):
    parts = []
    for col in merge_cols:
        val = row[col]
        if pd.notna(val):
            parts.append(f"{col}: {val}")
    return "; ".join(parts) if parts else None

pois_clean = pois[base_cols].copy()

pois_clean["additional_information"] = pois.apply(
    consolidate_info, axis=1
)

pois_clean.head()
```
