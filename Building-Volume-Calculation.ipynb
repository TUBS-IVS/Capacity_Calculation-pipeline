```python
gdf = info[info.geometry.notnull()].copy()
gdf = gdf.set_crs(epsg=4326, allow_override=True)

boundary_gdf = gpd.read_file(
    r"Areas-of-interest-POIs\regionalverband_area.gpkg"
)

boundary_gdf = boundary_gdf.to_crs(epsg=4326)
region_geom = boundary_gdf.geometry.union_all()

before_count = len(gdf)

filtered_gdf = gdf[gdf.geometry.intersects(region_geom)].copy()

after_count = len(filtered_gdf)

print("\n========== FILTER STATS ==========")
print(f"Total tiles before filtering : {before_count}")
print(f"Tiles after filtering        : {after_count}")
print(f"Tiles removed                : {before_count - after_count}")

percent_kept = (after_count / before_count) * 100 if before_count else 0
print(f"Percentage kept              : {percent_kept:.2f}%")

output_path = r"Areas-of-interest-POIs\filtered_tiles.geojson"
filtered_gdf.to_file(output_path, driver="GeoJSON")
```

```python
import requests
import zipfile
from pathlib import Path
from tqdm import tqdm

BASE_DIR = Path("Areas-of-interest-POIs")

COLLECTION_DIR = BASE_DIR / "Collection_folder"
UNZIP_DIR = BASE_DIR / "Unzipped-folder"
FINAL_SHP_DIR = BASE_DIR / "Final_folder"

COLLECTION_DIR.mkdir(parents=True, exist_ok=True)
UNZIP_DIR.mkdir(parents=True, exist_ok=True)
FINAL_SHP_DIR.mkdir(parents=True, exist_ok=True)

# ========== STATS ==========
download_success = 0
download_failed = 0
failed_urls = []

# ========== DOWNLOAD FUNCTION ==========
def download_file(url, dest_path, timeout=30):
    try:
        response = requests.get(url, stream=True, timeout=timeout)
        response.raise_for_status()

        with open(dest_path, "wb") as f:
            for chunk in response.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)
        return True
    except Exception as e:
        print(f" Failed: {url} | {e}")
        return False


# ========== STEP 1: DOWNLOAD ALL ZIPS ==========
print(" Downloading ZIP files...")

for idx, url in tqdm(enumerate(filtered_gdf["shp"]), total=len(filtered_gdf)):
    if not isinstance(url, str) or not url.startswith("http"):
        download_failed += 1
        failed_urls.append(url)
        continue

    file_name = f"file_{idx}.zip"
    dest_path = COLLECTION_DIR / file_name

    if dest_path.exists():
        download_success += 1
        continue

    success = download_file(url, dest_path)

    if success:
        download_success += 1
    else:
        download_failed += 1
        failed_urls.append(url)

# ========== STEP 2: UNZIP ALL FILES ==========
print(" Unzipping files...")

for zip_path in COLLECTION_DIR.glob("*.zip"):
    try:
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            extract_folder = UNZIP_DIR / zip_path.stem
            extract_folder.mkdir(exist_ok=True)
            zip_ref.extractall(extract_folder)
    except Exception as e:
        print(f" Failed to unzip {zip_path.name} | {e}")


# ========== FINAL STATS ==========
print("\n========== DOWNLOAD STATS ==========")
print(f"Total URLs         : {len(filtered_gdf)}")
print(f"Successful downloads: {download_success}")
print(f"Failed downloads    : {download_failed}")

if failed_urls:
    print("\n Failed URLs:")
    for url in failed_urls:
        print(url)
```

```python
from pathlib import Path

UNZIP_DIR = Path(r"Areas-of-interest-POIs")

shp_files = list(UNZIP_DIR.rglob("*.shp"))

list_file = UNZIP_DIR / "shp_list.txt"

with open(list_file, "w", encoding="utf-8") as f:
    for shp in shp_files:
        f.write(str(shp) + "\n")

print(f" Listed {len(shp_files)} shapefiles in {list_file}")
```

```python
import subprocess
from pathlib import Path

UNZIP_DIR = Path(r"Areas-of-interest-POIs")
output_gpkg = UNZIP_DIR / "merged_all.gpkg"
list_file = UNZIP_DIR / "shp_list.txt"

cmd = [
    "ogr2ogr",
    "-f", "GPKG",
    str(output_gpkg),
    "-nln", "buildings",
    "-update",
    "-append",
]

for line in open(list_file, "r", encoding="utf-8"):
    shp = line.strip()
    subprocess.run(cmd + [shp], check=True)

print(f" Successfully merged all shapefiles into {output_gpkg}")
```
