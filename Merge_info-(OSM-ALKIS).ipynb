```python
labels_to_remove = ['canopy',
                    'Agricultural and forestry business building',
                    'Buildings for supplying energy', 
                    'silo',
                    'mast',
                    'Solar cells',
                    'Wind turbine',
                    'Operational building for road traffic',
                    'tank',
                    'Radio mast',
                    'chimney',
                    'barracks',
                    'Transmission and radio tower',
                    'Creation',
                    'Observation tower',
                    'Water tower',
                    'Windmill',
                    'ski jump (inrun)']  # useless building types for our use case

gdf = gdf[~gdf["label_en"].apply(
    lambda x: any(label in labels_to_remove for label in (x if isinstance(x, list) else [x]))
)]
```

```python
import numpy as np

TARGET_CRS = 25832
DEFAULT_FLOOR_HEIGHT = 2.0
DEFAULT_FLOORS = 1

# keep ONLY the final schema you want
final_cols = [
    'gml_id', 'volume_m3', 'label_en', 'osm_names', 'osm_building_type',
    'osm_landuse_class', 'osm_landuse_name', 'gfk_class', 'gfk_name',
    'ALKIS_Landuse_info', 'alkis_address', 'geometry'
]

gdf_base = gdf[final_cols].copy()
osm_fill = osm_building_types.copy()

# keep only polygonal OSM buildings
osm_fill = osm_fill[osm_fill.geom_type.isin(["Polygon", "MultiPolygon"])].copy()

# keep only actual buildings
osm_fill = osm_fill[osm_fill["building"].notna()].copy()

# project both
gdf_base = gdf_base.to_crs(TARGET_CRS)
osm_fill = osm_fill.to_crs(TARGET_CRS)

# fix invalid geometries
gdf_base["geometry"] = gdf_base.geometry.buffer(0)
osm_fill["geometry"] = osm_fill.geometry.buffer(0)

# drop empty geometries
gdf_base = gdf_base[gdf_base.geometry.notna() & ~gdf_base.geometry.is_empty].copy()
osm_fill = osm_fill[osm_fill.geometry.notna() & ~osm_fill.geometry.is_empty].copy()

# find OSM buildings missing in ALKIS
join_check = gpd.sjoin(
    osm_fill,
    gdf_base[["geometry"]],
    how="left",
    predicate="intersects"
)

osm_missing = join_check[join_check["index_right"].isna()].copy()
osm_missing = osm_fill.loc[osm_missing.index].copy()

def to_float_safe(x):
    if pd.isna(x):
        return np.nan
    try:
        return float(str(x).replace(",", "."))
    except:
        return np.nan

def infer_height_m(row):
    h = to_float_safe(row.get("height"))
    if pd.notna(h):
        return h

    levels = to_float_safe(row.get("building:levels"))
    if pd.notna(levels):
        return levels * DEFAULT_FLOOR_HEIGHT

    return DEFAULT_FLOORS * DEFAULT_FLOOR_HEIGHT

def build_address(row):
    parts = [
        row.get("addr:street"),
        row.get("addr:housenumber"),
        row.get("addr:postcode"),
        row.get("addr:city"),
    ]
    parts = [str(p).strip() for p in parts if pd.notna(p) and str(p).strip() != ""]
    return ", ".join(parts) if parts else None

volume_m3 = osm_missing.geometry.area * osm_missing.apply(infer_height_m, axis=1)

# build ONLY the target schema
osm_to_add = pd.DataFrame(index=osm_missing.index)
osm_to_add["gml_id"] = "osm_" + osm_missing["id"].astype(str)
osm_to_add["volume_m3"] = volume_m3
osm_to_add["label_en"] = pd.NA
osm_to_add["osm_names"] = osm_missing["name"]               # put OSM name here
osm_to_add["osm_building_type"] = osm_missing["building"]
osm_to_add["osm_landuse_class"] = osm_missing["landuse"]
osm_to_add["osm_landuse_name"] = osm_missing["landuse"]
osm_to_add["gfk_class"] = pd.NA
osm_to_add["gfk_name"] = pd.NA
osm_to_add["ALKIS_Landuse_info"] = pd.NA
osm_to_add["alkis_address"] = osm_missing.apply(build_address, axis=1)
osm_to_add["geometry"] = osm_missing.geometry

osm_to_add = gpd.GeoDataFrame(osm_to_add, geometry="geometry", crs=TARGET_CRS)

# force exact same columns before concat
osm_to_add = osm_to_add[final_cols]
gdf_base = gdf_base[final_cols]

gdf = pd.concat([gdf_base, osm_to_add], ignore_index=True)
gdf = gpd.GeoDataFrame(gdf, geometry="geometry", crs=TARGET_CRS)

print("ALKIS buildings:", len(gdf_base))
print("Added OSM-only buildings:", len(osm_to_add))
print("Final total buildings:", len(gdf))
```

```python
def clean_join_cols(df):
    return df.drop(columns=["index_right", "index_left"], errors="ignore")

gdf = clean_join_cols(gdf.to_crs(25832).copy())
clean_pois = clean_join_cols(clean_pois.to_crs(25832).copy())

gdf["gml_id"] = gdf.index

# direct intersect
pois_joined = gpd.sjoin(
    clean_join_cols(clean_pois),
    gdf[["gml_id", "geometry"]],
    how="left",
    predicate="intersects"
)

pois_joined["match_type"] = pois_joined["gml_id"].notna().map(
    {True: "intersects", False: None}
)

matched = pois_joined[pois_joined["gml_id"].notna()].copy()
unmatched_initial = pois_joined[pois_joined["gml_id"].isna()].copy()

# nearest fallback for non-intersecting POIs
if not unmatched_initial.empty:
    nearest = gpd.sjoin_nearest(
        clean_join_cols(unmatched_initial.drop(columns=["gml_id"], errors="ignore")),
        gdf[["gml_id", "geometry"]],
        how="left",
        max_distance=100,
        distance_col="snap_dist"
    )

    nearest["match_type"] = nearest["gml_id"].notna().map(
        {True: "nearest", False: "unmatched"}
    )

    pois_final = pd.concat([matched, nearest], ignore_index=True)

    # only POIs that still found no building even after nearest
    unmatched_final = nearest[nearest["gml_id"].isna()].copy()

else:
    pois_final = matched.copy()
    unmatched_final = unmatched_initial.copy()

# merge back to buildings
result = gdf.merge(
    pois_final.drop(columns="geometry"),
    on="gml_id",
    how="left"
)

# result.to_file("output.gpkg", driver="GPKG")
# unmatched_final.to_file("unmatched_pois.gpkg", driver="GPKG")
```

```python
def is_missing(x):
    if x is None:
        return True
    try:
        return pd.isna(x)
    except Exception:
        return False

def flatten_values(series):
    """
    Flatten mixed values from a series into a simple python list.
    Handles scalars, lists, tuples, sets.
    """
    out = []
    for v in series:
        if is_missing(v):
            continue
        if isinstance(v, (list, tuple, set)):
            for item in v:
                if not is_missing(item):
                    out.append(item)
        else:
            out.append(v)
    return out

def unique_preserve_order(values):
    seen = set()
    out = []
    for v in values:
        key = str(v).strip()
        if key == "":
            continue
        if key not in seen:
            seen.add(key)
            out.append(v)
    return out

def first_non_null(series):
    for v in series:
        if not is_missing(v):
            return v
    return None

def merge_to_unique_list(series):
    vals = flatten_values(series)
    vals = [str(v).strip() for v in vals if not is_missing(v) and str(v).strip() != ""]
    vals = unique_preserve_order(vals)
    return vals if vals else None

def merge_names(osm_names_series, name_series):
    """
    Combine existing osm_names + name into one unique list.
    """
    vals = []

    # existing osm_names may already be string/list
    for v in osm_names_series:
        if is_missing(v):
            continue
        if isinstance(v, (list, tuple, set)):
            vals.extend([str(x).strip() for x in v if not is_missing(x) and str(x).strip() != ""])
        else:
            vals.append(str(v).strip())

    # add name column too
    for v in name_series:
        if is_missing(v):
            continue
        vals.append(str(v).strip())

    vals = [v for v in vals if v != ""]
    vals = unique_preserve_order(vals)
    return vals if vals else None

def merge_address(alkis_series, addrfull_series):
    """
    Keep one address field by combining alkis_address and addr:full.
    If both exist and differ, keep both in a single string separated by ' | '.
    """
    vals = []

    for v in alkis_series:
        if not is_missing(v):
            s = str(v).strip()
            if s:
                vals.append(s)

    for v in addrfull_series:
        if not is_missing(v):
            s = str(v).strip()
            if s:
                vals.append(s)

    vals = unique_preserve_order(vals)

    if not vals:
        return None
    if len(vals) == 1:
        return vals[0]

    return " | ".join(vals)

def keep_single_or_first(series):
    """
    For repeated building-level columns.
    If values are repeated, keeps one.
    If not perfectly repeated, keeps first non-null.
    """
    non_null = [v for v in series if not is_missing(v)]
    if not non_null:
        return None
    return non_null[0]


# ---------- columns ----------
single_value_cols = [
    "volume_m3",
    "label_en",
    "osm_building_type",
    "osm_landuse_class",
    "osm_landuse_name",
    "gfk_class",
    "gfk_name",
    "ALKIS_Landuse_info",
    "geometry"
]

list_cols = [
    "email",
    "amenity",
    "building",
    "shop",
    "tourism",
    "information",
    "tags_search",
    "additional_information",
]

def min_non_null(series):
    vals = [v for v in series if not is_missing(v)]
    return min(vals) if vals else None

grouped = result.groupby("gml_id", dropna=False)

rows = []

for gml_id, grp in grouped:
    row = {"gml_id": gml_id}

    # building-level repeated columns
    for col in single_value_cols:
        if col in grp.columns:
            row[col] = keep_single_or_first(grp[col])

    # merge names into one osm_names field
    row["osm_names"] = merge_names(
        grp["osm_names"] if "osm_names" in grp.columns else pd.Series(dtype=object),
        grp["name"] if "name" in grp.columns else pd.Series(dtype=object)
    )

    # merge address fields into one
    row["alkis_address"] = merge_address(
        grp["alkis_address"] if "alkis_address" in grp.columns else pd.Series(dtype=object),
        grp["addr:full"] if "addr:full" in grp.columns else pd.Series(dtype=object)
    )

    # business-level fields as unique lists
    for col in list_cols:
        if col in grp.columns:
            row[col] = merge_to_unique_list(grp[col])

    # optional: keep unique websites too
    if "website" in grp.columns:
        row["website"] = merge_to_unique_list(grp["website"])

    rows.append(row)

condensed = gpd.GeoDataFrame(rows, geometry="geometry", crs=result.crs)
```
