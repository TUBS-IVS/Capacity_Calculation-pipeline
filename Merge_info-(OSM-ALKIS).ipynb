```python
import geopandas as gpd
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.width", 0)

gdf = gpd.read_file('Areas-of-interest-POIs/Enriched-building-volume-data.gpkg')
print(gdf.crs)
```

```python
gdf.info()
```

```python
gdf = gdf.drop(['measHeight', 'function', 'area_m2'],axis=1)
```

```python
gdf.head()
```

```python
gdf["alkis_address"] = (
    gdf[["Name", "Strasse", "HausNr", "Stadt"]]
    .apply(
        lambda row: " ".join(
            [str(v) for v in row if pd.notna(v) and str(v).strip() != ""]
        ),
        axis=1
    )
)

gdf = gdf.drop(["Name", "Strasse", "HausNr", "Stadt"], axis=1)

gdf.head()
```

```python
gdf = gdf[['gml_id','volume_m3','label_en','osm_names','osm_building_type','osm_landuse_class','osm_landuse_name','gfk_class','gfk_name','ALKIS_Landuse_info','alkis_address','geometry']]
gdf.head()
```

```python
gdf.info()
```

```python
osm_pois = gpd.read_file('Areas-of-interest-POIs/OSM-POIs-modified.gpkg')
osm_pois.head()
```

```python
osm_pois.info()
```

```python
# ensure same CRS
if gdf.crs != osm_pois.crs:
    osm_pois = osm_pois.to_crs(gdf.crs)
```

```python
joined = gpd.sjoin(
    gdf,
    osm_pois,
    how="left",
    predicate="intersects"
)

# ensure proper merge (check if here the pois which outside the polygons are merged properly!)
```

```python
out_cols = {
    # IDs
    "id": "osm_id",
    "gml_id": "gml_id",

    # names
    "name": "name",
    "osm_names": "osm_names",

    # contact / web
    "email": "email",
    "website": "website",

    # building attributes
    "volume_m3": "volume_m3",
    "osm_building_type": "osm_building_type",
    "osm_landuse_class": "osm_landuse_class",
    "osm_landuse_name": "osm_landuse_name",
    "gfk_class": "gfk_class",
    "gfk_name": "gfk_name",
    "ALKIS_Landuse_info": "ALKIS_Landuse_info",
    "alkis_address": "alkis_address",

    # OSM tags
    "amenity": "amenity",
    "building": "building",
    "shop": "shop",
    "tourism": "tourism",
    "information": "information",

    # text blobs
    "addr:full": "addr:full",
    "tags_search": "tags_search",
    "additional_information": "additional_information",

    # geometry (from gdf!)
    "geometry": "geometry"
}

result = joined[list(out_cols.keys())].rename(columns=out_cols)

result = gpd.GeoDataFrame(result, geometry="geometry", crs=gdf.crs)
result.head()
```

```python
result.info()   
```

```python
agg_rules = {
    "osm_id": lambda x: list(x.dropna().unique()),
    "name": lambda x: "; ".join(x.dropna().unique()),
    "amenity": lambda x: "; ".join(x.dropna().unique()),
    "shop": lambda x: "; ".join(x.dropna().unique()),
    "tourism": lambda x: "; ".join(x.dropna().unique()),
    "information": lambda x: "; ".join(x.dropna().unique()),
    "tags_search": lambda x: "; ".join(x.dropna().unique()),
    "additional_information": lambda x: "; ".join(x.dropna().unique()),

    # keep building attrs stable
    "volume_m3": "first",
    "osm_building_type": "first",
    "osm_landuse_class": "first",
    "osm_landuse_name": "first",
    "gfk_class": "first",
    "gfk_name": "first",
    "ALKIS_Landuse_info": "first",
    "alkis_address": "first",
    "geometry": "first",
}

result_agg = (
    result
    .groupby("gml_id", as_index=False)
    .agg(agg_rules)
)

result_agg = gpd.GeoDataFrame(result_agg, geometry="geometry", crs=gdf.crs)
result_agg.head()
```
