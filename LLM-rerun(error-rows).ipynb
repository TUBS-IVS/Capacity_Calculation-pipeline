```python
# -----------------------------
# MiD labels (STRICT)
# -----------------------------

TARGET_MID_LABELS = {
    "work",
    "university",
    "school",
    "childcare",
    "retail_daily",
    "retail_non_daily",
    "leisure",
    "sports",
    "errands",
    "meetup",
    "lessons",
    "business"
}

# -----------------------------
# JSON extraction
# -----------------------------
def extract_first_json_object(text: str) -> dict:
    """
    Extract the first JSON object from model output.
    """
    m = re.search(r"\{.*\}", text, flags=re.DOTALL)
    if not m:
        raise ValueError("No JSON object found in output.")
    return json.loads(m.group(0))


# -----------------------------
# Validation (PROMPT-ALIGNED)
# -----------------------------
def validate(obj: dict):
    # ---- Required keys ----
    if "interpreted_type" not in obj or not isinstance(obj["interpreted_type"], str):
        raise ValueError("Missing or invalid interpreted_type")

    if "mid_labels" not in obj or not isinstance(obj["mid_labels"], list):
        raise ValueError("mid_labels must be a list")

    if "bosserhof_class" not in obj:
        raise ValueError("Missing bosserhof_class")

    if "reason" not in obj or not isinstance(obj["reason"], str):
        raise ValueError("Missing or invalid reason")

    # ---- Validate MiD labels ----
    for lab in obj["mid_labels"]:
        if lab not in TARGET_MID_LABELS:
            raise ValueError(f"Invalid MiD label: {lab}")

    # ---- Validate Bosserhof class ----
    bc = obj["bosserhof_class"]

    # Allowed: no assignment
    if bc is None or bc == []:
        pass

    # Allowed: any non-empty string (granular OR fallback)
    elif isinstance(bc, str) and bc.strip():
        pass

    else:
        raise ValueError("bosserhof_class must be a non-empty string, [] or None")

    # ---- Consistency rule ----
    # If there is no building-related visitor activity,
    # there must not be a Bosserhof assignment
    if len(obj["mid_labels"]) == 0 and bc not in (None, []):
        raise ValueError(
            "bosserhof_class assigned but mid_labels is empty — inconsistent"
        )
```

```python
def call_tu_llm(
    user_input: str = None,
    batch_inputs: list = None,
    max_retries=3,
    backoff_sec=2.0,
    debug=False,
) -> str:

    headers = {
        "Authorization": f"Bearer {TU_TOKEN}",
        "Accept": "application/json",
        "Content-Type": "application/json",
    }

    # -----------------------------
    # Build prompt
    # -----------------------------
    if batch_inputs is not None:
        items = []
        for i, row in enumerate(batch_inputs, start=1):
            gml_id = str(row["gml_id"])
            sentence = "" if row.get("sentence") is None else str(row.get("sentence"))
            items.append(
                f"ITEM {i}\n"
                f"gml_id: {gml_id}\n"
                f"sentence: {sentence}"
            )
        prompt_text = "\n\n".join(items)
    else:
        prompt_text = "" if user_input is None else str(user_input)

    payload = {
        "thread": None,
        "prompt": prompt_text,
        "model": MODEL,
        "customInstructions": SYSTEM_PROMPT,
        "hideCustomInstructions": True,
    }

    last_err = None

    for attempt in range(1, max_retries + 1):
        try:
            r = requests.post(
                API_URL,
                headers=headers,
                json=payload,
                stream=True,
                timeout=120,
            )

            if debug:
                print("STATUS:", r.status_code)
                try:
                    print("REQUEST PAYLOAD:", json.dumps(payload, indent=2)[:4000])
                except Exception:
                    pass

            r.raise_for_status()

            full_text = ""

            for line in r.iter_lines(decode_unicode=True):
                if not line:
                    continue
                try:
                    event = json.loads(line)
                except json.JSONDecodeError:
                    continue

                if event.get("type") == "chunk":
                    full_text += event.get("content", "")
                elif event.get("type") == "done":
                    if "response" in event:
                        full_text = event["response"]
                    break

            return full_text

        except Exception as e:
            last_err = e
            if debug:
                print(f"Attempt {attempt} failed: {e}")
            time.sleep(backoff_sec * attempt)

    raise RuntimeError(f"TU LLM call failed: {last_err}")
```

```python
def predict_from_sentence(gml_id=None, sentence=None, batch_rows=None, debug=False):
    # -------------------------------------------------
    # BATCH MODE
    # -------------------------------------------------
    if batch_rows is not None:
        try:
            raw = call_tu_llm(batch_inputs=batch_rows, debug=debug)
            objs = extract_first_json_array(raw)

            if not isinstance(objs, list):
                raise ValueError("Batch output is not a list.")

            if len(objs) != len(batch_rows):
                raise ValueError(
                    f"Batch size mismatch: got {len(objs)} outputs for {len(batch_rows)} inputs."
                )

            results = []
            for row, obj in zip(batch_rows, objs):
                row_gml_id = row["gml_id"]

                try:
                    validate(obj)
                    results.append({
                        "gml_id": row_gml_id,
                        "interpreted_type": obj["interpreted_type"],
                        "mid_labels": obj["mid_labels"],
                        "bosserhof_class": obj["bosserhof_class"],
                    })
                except Exception:
                    results.append({
                        "gml_id": row_gml_id,
                        "interpreted_type": "error",
                        "mid_labels": [],
                        "bosserhof_class": None,
                    })

            return results

        except Exception:
            return [
                {
                    "gml_id": row["gml_id"],
                    "interpreted_type": "error",
                    "mid_labels": [],
                    "bosserhof_class": None,
                }
                for row in batch_rows
            ]

    # -------------------------------------------------
    # SINGLE MODE
    # -------------------------------------------------
    gml_id = gml_id.item() if hasattr(gml_id, "item") else gml_id
    sentence = "" if sentence is None else str(sentence)

    try:
        raw = call_tu_llm(user_input=sentence, debug=debug)
        obj = extract_first_json_object(raw)
        validate(obj)

        return {
            "gml_id": gml_id,
            "interpreted_type": obj["interpreted_type"],
            "mid_labels": obj["mid_labels"],
            "bosserhof_class": obj["bosserhof_class"],
        }

    except Exception:
        return {
            "gml_id": gml_id,
            "interpreted_type": "error",
            "mid_labels": [],
            "bosserhof_class": None,
        }
    
def extract_first_json_array(text: str):
    import json
    import re

    match = re.search(r"\[\s*{.*}\s*\]", text, flags=re.DOTALL)
    if not match:
        raise ValueError("No JSON array found in model response.")
    return json.loads(match.group(0))
```

```python
import time
import math
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor, as_completed

import pandas as pd


# ---------------------------------------------------------
# Helper: split list into chunks
# ---------------------------------------------------------
def chunk_list(items, batch_size):
    for i in range(0, len(items), batch_size):
        yield items[i:i + batch_size]


# ---------------------------------------------------------
# RETRY WRAPPER FOR A WHOLE BATCH
# ---------------------------------------------------------
def predict_batch_with_retry(batch_rows, debug=False, max_retries=3, wait_sec=5):
    last_err = None
    batch_ids = [str(r["gml_id"]) for r in batch_rows]

    for attempt in range(1, max_retries + 1):
        try:
            return predict_batch_from_sentences(
                batch_rows=batch_rows,
                debug=debug,
            )
        except Exception as e:
            last_err = e
            print(f"[batch {batch_ids[:3]}... size={len(batch_rows)}] attempt {attempt}/{max_retries} failed: {e}")
            time.sleep(wait_sec)

    raise last_err


# ---------------------------------------------------------
# THIS IS THE MAIN PLACE TO IMPLEMENT REAL API BATCHING
# ---------------------------------------------------------
def predict_batch_from_sentences(batch_rows, debug=False):
    results = []

    prompts = []
    for row in batch_rows:
        gml_id = str(row["gml_id"])
        sentence = row.get("sentence", "")
        prompts.append({
            "gml_id": gml_id,
            "sentence": sentence,
        })

    res = predict_from_sentence(
        batch_rows=prompts,
        debug=debug,
    )

    if not isinstance(res, list):
        raise ValueError("Batch prediction must return a list.")

    if len(res) != len(batch_rows):
        raise ValueError(
            f"Batch output length mismatch: got {len(res)} results for {len(batch_rows)} rows."
        )

    for item in res:
        if isinstance(item, dict):
            results.append(item)
        else:
            results.append({"prediction": item})

    return results


# ---------------------------------------------------------
# APPEND TO PARQUET
# ---------------------------------------------------------
def append_parquet(path, new_rows, deduplicate_on=None):
    if not new_rows:
        return

    new_df = pd.DataFrame(new_rows)

    if path.exists():
        old_df = pd.read_parquet(path)
        merged = pd.concat([old_df, new_df], ignore_index=True)
    else:
        merged = new_df.copy()

    if deduplicate_on is not None:
        merged = merged.drop_duplicates(subset=[deduplicate_on], keep="last")

    merged.to_parquet(path, index=False)


# ---------------------------------------------------------
# MAIN FUNCTION: PARALLEL + BATCHED
# ---------------------------------------------------------
def classify_resume_parallel_batched(
    pois_df: pd.DataFrame,
    checkpoint_path="predictions_checkpoint.parquet",
    error_path="prediction_errors.parquet",
    save_every=200,
    max_workers=5,
    batch_size=10,
    debug=False,
    max_retries=3,
    retry_wait_sec=5,
):
    checkpoint_path = Path(checkpoint_path)
    error_path = Path(error_path)

    df = pois_df.copy().reset_index(drop=True)

    if "gml_id" not in df.columns:
        raise ValueError("pois_df must contain a 'gml_id' column.")

    if "sentence" not in df.columns:
        raise ValueError("pois_df must contain a 'sentence' column.")

    df["gml_id"] = df["gml_id"].astype(str)

    # do not checkpoint geometry
    cols_to_save = [c for c in df.columns if c != "geometry"]

    # ---------------------------
    # Load checkpoint
    # ---------------------------
    if checkpoint_path.exists():
        checkpoint_df = pd.read_parquet(checkpoint_path)
        checkpoint_df["gml_id"] = checkpoint_df["gml_id"].astype(str)
        done_ids = set(checkpoint_df["gml_id"])
        print(f"Loaded checkpoint with {len(done_ids)} completed rows")
    else:
        done_ids = set()
        print("No checkpoint found. Starting from scratch.")

    remaining_df = df[~df["gml_id"].isin(done_ids)].copy()
    print(f"Remaining rows to process: {len(remaining_df)} / {len(df)}")

    records = remaining_df.to_dict("records")
    batches = list(chunk_list(records, batch_size))

    print(f"Total batches to process: {len(batches)} | batch_size={batch_size} | max_workers={max_workers}")

    def batch_worker(batch_rows):
        batch_results = predict_batch_with_retry(
            batch_rows=batch_rows,
            debug=debug,
            max_retries=max_retries,
            wait_sec=retry_wait_sec,
        )

        if not isinstance(batch_results, list):
            raise ValueError("predict_batch_from_sentences() must return a list.")

        if len(batch_results) != len(batch_rows):
            raise ValueError(
                f"Batch output length mismatch: got {len(batch_results)} results for {len(batch_rows)} rows."
            )

        out_rows = []
        for row_dict, res in zip(batch_rows, batch_results):
            out_row = {k: v for k, v in row_dict.items() if k in cols_to_save}

            if isinstance(res, dict):
                out_row.update(res)
            else:
                out_row["prediction"] = res

            out_rows.append(out_row)

        return out_rows

    buffer = []
    error_buffer = []
    processed_now = 0
    completed_batches = 0

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        future_to_batch = {
            executor.submit(batch_worker, batch_rows): batch_rows
            for batch_rows in batches
        }

        for future in as_completed(future_to_batch):
            batch_rows = future_to_batch[future]
            batch_ids = [str(r["gml_id"]) for r in batch_rows]

            try:
                out_rows = future.result()
                buffer.extend(out_rows)

                processed_now += len(out_rows)
                completed_batches += 1

                print(
                    f"Processed {processed_now}/{len(remaining_df)} new rows "
                    f"| completed batches {completed_batches}/{len(batches)} "
                    f"| last batch size={len(batch_rows)} "
                    f"| gml_id range={batch_ids[0]} -> {batch_ids[-1]}"
                )

            except Exception as e:
                for row in batch_rows:
                    error_buffer.append({
                        "gml_id": str(row["gml_id"]),
                        "error": str(e),
                    })

                completed_batches += 1
                print(
                    f"Failed batch {completed_batches}/{len(batches)} "
                    f"| size={len(batch_rows)} "
                    f"| gml_id range={batch_ids[0]} -> {batch_ids[-1]} "
                    f"| error={e}"
                )

            # Save regularly from main thread only
            if len(buffer) >= save_every:
                append_parquet(checkpoint_path, buffer, deduplicate_on="gml_id")
                print(f"Checkpoint saved with {len(buffer)} new rows")
                buffer = []

            if len(error_buffer) >= save_every:
                append_parquet(error_path, error_buffer, deduplicate_on="gml_id")
                print(f"Error log saved with {len(error_buffer)} rows")
                error_buffer = []

    # final flush
    if buffer:
        append_parquet(checkpoint_path, buffer, deduplicate_on="gml_id")
        print("Final checkpoint saved.")

    if error_buffer:
        append_parquet(error_path, error_buffer, deduplicate_on="gml_id")
        print("Final error log saved.")

    final_df = pd.read_parquet(checkpoint_path)
    final_df["gml_id"] = final_df["gml_id"].astype(str)

    return final_df
```
